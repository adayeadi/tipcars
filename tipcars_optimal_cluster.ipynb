```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
data = pd.read_csv('tipcars_preprocessed.csv', delimiter=';')
```

```python
# Nastavení dummy pro atribut stav
data_cluster = data.drop(columns=['id','barva', 'metaliza', 'typ', 'znacka', 'model', 'palivo', 'karoserie'])
data_cluster = pd.get_dummies(data_cluster, columns=['stav'])

data_cluster['objem'] = data_cluster['objem'].fillna(0)
```

```python
# Hledání optimálního počtu shluků pomocí Elbow metody a Silhouette koeficientu
range_n_clusters = range(2, 20)

silhouette_scores = []
inertia = []

for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(data_cluster)
    
    silhouette_avg = silhouette_score(data_cluster, cluster_labels)
    silhouette_scores.append(silhouette_avg)
    print(f"Pro n_clusters = {n_clusters}, průměrný siluetový koeficient = {silhouette_avg:.2f}")
    
    inertia.append(kmeans.inertia_)
```

```python
#Vizualizace Silhuetového koeficientu
plt.figure(figsize=(12, 5))
plt.plot(range_n_clusters, silhouette_scores, marker='o')
plt.xlabel('Počet shluků')
plt.ylabel('Silhuetový koeficient')
plt.title('Silhuetový koeficient pro optimální počet shluků')
plt.show()
```

```python
#Vizualizace Elbow metody
plt.figure(figsize=(12, 5))
plt.plot(range_n_clusters, inertia, marker='o')
plt.xlabel('Počet shluků')
plt.ylabel('Inercie')
plt.title('Elbow metoda pro optimální počet shluků')
plt.show()
```
